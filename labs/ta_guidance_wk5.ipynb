### Business Question 1: Households with Children and Snack Purchases

**Teaching Point**: This demonstrates categorical comparison visualization

```python
# Complete solution for Challenge 1
import pandas as pd
import matplotlib.pyplot as plt
from completejourney_py import get_data

# Load data
cj = get_data()
transactions = cj['transactions']
products = cj['products']
demographics = cj['demographics']

# The analysis code from the lab (already provided)
tx_prod = transactions.merge(products, on="product_id", how="left")
snack_sales = tx_prod[tx_prod['product_type'].str.contains('Snacks', case=False, na=False)]
snack_demo = snack_sales.merge(demographics, on="household_id", how="left")
snack_demo['has_kids'] = snack_demo['kids_count'] != '0'
sales_by_children = (
    snack_demo
    .groupby("has_kids", as_index=False)
    .agg(total_sales=("sales_value", "sum"))
)

# Solution 1: Simple bar chart with pandas
ax = sales_by_children.set_index('has_kids')['total_sales'].plot(kind='bar', color=['skyblue', 'lightcoral'])
ax.set_title('Snack Sales: Households With vs Without Children')
ax.set_xlabel('Has Children')
ax.set_ylabel('Total Sales ($)')
ax.set_xticklabels(['No Children', 'Has Children'], rotation=0)
plt.tight_layout()
plt.show()

# Solution 2: Enhanced matplotlib version
fig, ax = plt.subplots(figsize=(8, 6))
categories = ['No Children', 'Has Children']
values = sales_by_children['total_sales'].values

bars = ax.bar(categories, values, color=['#3498db', '#e74c3c'], alpha=0.8)
ax.set_title('Snack Sales by Household Type', fontsize=14, fontweight='bold')
ax.set_ylabel('Total Sales ($)', fontsize=12)
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'${x:,.0f}'))

# Add value labels on bars
for bar, value in zip(bars, values):
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2., height + height*0.01,
            f'${value:,.0f}', ha='center', va='bottom', fontweight='bold')

plt.tight_layout()
plt.show()

# Business insight
print("Business Insight:")
if values[1] > values[0]:
    difference = values[1] - values[0]
    percent_more = (difference / values[0]) * 100
    print(f"Households with children spend ${difference:,.0f} more on snacks ({percent_more:.1f}% increase)")
    print("Recommendation: Target snack marketing toward families with children")
else:
    print("Households without children actually spend more on snacks - investigate why")
```

### Business Question 2: December 2017 Product Categories

**Teaching Point**: This demonstrates ranking/top-N visualization

```python
# Complete solution for Challenge 2
# First, recreate the analysis from the lab notebook

# Filter to December 2017
december_tx = transactions[
    (transactions["transaction_timestamp"].dt.year == 2017) &
    (transactions["transaction_timestamp"].dt.month == 12)
]

# Join products for category
december_tx_prod = december_tx.merge(products, on="product_id", how="left")

# Aggregate sales by category
sales_by_category_dec = (
    december_tx_prod
    .groupby("product_category", as_index=False)
    .agg(total_sales=("sales_value", "sum"))
    .sort_values("total_sales", ascending=False)
)

# Solution 1: Top 10 horizontal bar chart
top_10_categories = sales_by_category_dec.head(10)

plt.figure(figsize=(10, 6))
ax = plt.barh(range(len(top_10_categories)), top_10_categories['total_sales'], 
              color='forestgreen', alpha=0.7)
plt.yticks(range(len(top_10_categories)), top_10_categories['product_category'])
plt.xlabel('Total Sales ($)')
plt.title('Top 10 Product Categories - December 2017 Sales', fontweight='bold')
plt.gca().invert_yaxis()  # Highest at top

# Add value labels
for i, v in enumerate(top_10_categories['total_sales']):
    plt.text(v + max(top_10_categories['total_sales'])*0.01, i, f'${v:,.0f}', 
             va='center', fontweight='bold')

plt.tight_layout()
plt.show()

# Solution 2: Interactive bokeh version
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import HoverTool

output_notebook()

p = figure(
    y_range=list(top_10_categories['product_category'][::-1]),  # Reverse for top-to-bottom
    height=400,
    title="Top 10 Categories - December 2017 (Interactive)",
    toolbar_location=None
)

p.hbar(y=list(top_10_categories['product_category'][::-1]), 
       right=list(top_10_categories['total_sales'][::-1]), 
       height=0.6, color='forestgreen', alpha=0.7)

hover = HoverTool(tooltips=[("Category", "@y"), ("Sales", "@right{$0,0}")])
p.add_tools(hover)
p.xaxis.axis_label = "Sales ($)"
show(p)

# Business insight
print("\nBusiness Insights:")
print(f"1. Top category: {top_10_categories.iloc[0]['product_category']} (${top_10_categories.iloc[0]['total_sales']:,.0f})")
print(f"2. Bottom of top 10: {top_10_categories.iloc[9]['product_category']} (${top_10_categories.iloc[9]['total_sales']:,.0f})")
print("3. Consider holiday promotions for top-performing December categories")
```

### Business Question 3: Soft Drink Trends Over Time

**Teaching Point**: This demonstrates time series visualization

```python
# Complete solution for Challenge 3
# First, recreate the analysis from the lab notebook

# Join products for category
tx_prod = transactions.merge(products[["product_id", "product_category"]],
                             on="product_id", how="left")

# Filter to soft drinks
soft_drinks = tx_prod[tx_prod["product_category"].str.contains("Soft drinks", case=False, na=False)].copy()

# Aggregate quantity by month
soft_drinks_monthly = (
    soft_drinks
    .set_index("transaction_timestamp")
    .resample("ME")["quantity"]
    .sum()
    .reset_index()
)

# Solution 1: Matplotlib line chart with trend
plt.figure(figsize=(12, 6))
plt.plot(soft_drinks_monthly['transaction_timestamp'], 
         soft_drinks_monthly['quantity'], 
         marker='o', linewidth=2, markersize=6, color='dodgerblue')

plt.title('Soft Drink Quantity Trends Over Time', fontsize=14, fontweight='bold')
plt.xlabel('Month', fontsize=12)
plt.ylabel('Total Quantity Sold', fontsize=12)
plt.xticks(rotation=45)
plt.grid(True, alpha=0.3)

# Add trend line
import numpy as np
from scipy import stats

# Convert dates to numbers for trend calculation
x_numeric = np.arange(len(soft_drinks_monthly))
slope, intercept, r_value, p_value, std_err = stats.linregress(x_numeric, soft_drinks_monthly['quantity'])
trend_line = slope * x_numeric + intercept

plt.plot(soft_drinks_monthly['transaction_timestamp'], trend_line, 
         '--', color='red', linewidth=2, alpha=0.8, label=f'Trend (slope: {slope:.1f})')
plt.legend()
plt.tight_layout()
plt.show()

# Solution 2: Seasonal analysis
# Add month and year columns for seasonal analysis
soft_drinks_analysis = soft_drinks_monthly.copy()
soft_drinks_analysis['month'] = soft_drinks_analysis['transaction_timestamp'].dt.month
soft_drinks_analysis['year'] = soft_drinks_analysis['transaction_timestamp'].dt.year

# Monthly averages
monthly_avg = soft_drinks_analysis.groupby('month')['quantity'].mean().reset_index()

plt.figure(figsize=(10, 6))
month_names = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 
               'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
plt.bar(range(1, 13), monthly_avg['quantity'], color='lightcoral', alpha=0.7)
plt.xticks(range(1, 13), month_names)
plt.title('Average Soft Drink Sales by Month', fontweight='bold')
plt.xlabel('Month')
plt.ylabel('Average Quantity')
plt.grid(True, alpha=0.3, axis='y')
plt.tight_layout()
plt.show()

# Business insights
print("\nBusiness Insights:")
if slope > 0:
    print(f"âœ… Positive trend: Soft drink sales increasing by {slope:.1f} units per month on average")
elif slope < 0:
    print(f"âš ï¸ Negative trend: Soft drink sales decreasing by {abs(slope):.1f} units per month")
else:
    print("â†’ Flat trend: Soft drink sales remaining stable")

peak_month = monthly_avg.loc[monthly_avg['quantity'].idxmax(), 'month']
low_month = monthly_avg.loc[monthly_avg['quantity'].idxmin(), 'month']
print(f"ðŸ“ˆ Peak month: {month_names[peak_month-1]}")
print(f"ðŸ“‰ Lowest month: {month_names[low_month-1]}")
print("ðŸ’¡ Consider seasonal promotions based on these patterns")
```
